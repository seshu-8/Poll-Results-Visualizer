# 📊 Poll Results Visualizer - EDA Notebook
Exploratory Data Analysis on the synthetic poll dataset.

```python
import sys
sys.path.insert(0, '../src')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from generate_data import generate_poll_data
from clean_data import clean_data
from analyze import overall_distribution, demographic_breakdown, summary_table

# Generate + clean
df_raw, questions = generate_poll_data(500)
df = clean_data(df_raw)
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
# ── Distribution of Q1 ──
dist_q1 = overall_distribution(df, 'Q1')
print(dist_q1)

plt.figure(figsize=(8, 4))
plt.barh(dist_q1['option'], dist_q1['votes'], color='steelblue')
plt.title('Q1: Programming Language Preference')
plt.xlabel('Votes')
plt.tight_layout()
plt.show()
```

```python
# ── Region Breakdown for Q1 ──
region_ct = demographic_breakdown(df, 'Q1', 'region')
region_ct.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Region-wise: Q1 Programming Language Preference')
plt.ylabel('%')
plt.xticks(rotation=30)
plt.legend(bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

```python
# ── Summary Table ──
summary = summary_table(df, questions)
summary
```

```python
# ── Monthly Trend ──
from analyze import trend_over_time
trend = trend_over_time(df, 'Q1', top_n=3)
trend.plot(figsize=(12, 4), marker='o')
plt.title('Monthly Trend: Q1')
plt.xlabel('Month')
plt.ylabel('Responses')
plt.tight_layout()
plt.show()
```

```python
# ── Heatmap: Region vs Q1 ──
region_ct = demographic_breakdown(df, 'Q1', 'region')
plt.figure(figsize=(10, 4))
sns.heatmap(region_ct, annot=True, fmt='.1f', cmap='YlOrRd')
plt.title('Heatmap: Region vs Q1 Options (%)')
plt.tight_layout()
plt.show()
```
